**Question 1.**

For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.  
1. The sample size n is extremely large, and the number of predic- tors p is small.
2. The number of predictors p is extremely large, and the number of observations n is small.
3. The relationship between the predictors and response is highly non-linear.
4. The variance of the error terms, i.e. σ2 = Var(ε), is extremely high.

**Solution.** 

$$\begin{align*}
\mathrm{E}\Big[\big(Y - \hat{f}(x)\big)^2 \Big|\, X=x_0 \Big]
 & =  \sigma_\varepsilon^2 + \mathrm{Bias}^2\big[\hat{f}(x_0)\big] + \mathrm{Var}\left[ \hat{f}(x_0) \right] , \
\end{align*}$$

$$
\begin{align*}
 \mathrm{Bias}\big[\hat{f}(x_0)\big] = \mathrm{E}\big[\hat{f}(x_0) - f(x_0)\big],
\end{align*}
$$

$$
\begin{align*}
\mathrm{Var}\big[\hat{f}(x_0)\big] = \mathrm{E}[\hat{f}(x_0)^2] - \mathrm{E}[\hat{f}(x_0)]^2.
\end{align*}
$$

1. Flexible. Flexible learning methods tend to do better when there is plenty of data and fewer predictors, lessening the chances of over learning. At this stage, the improvements made in lowering bias outperformed the increase in variance of the predicted model.  
2. Inflexible. In cases where there is scarce data and many predictors, flexible learning methods are prone to overlearning.    
3. Flexible. Highly non linear data is hard to approximate using simple, linear learning methods, which introduce greater levels of bias.   
4. The error term is part of the irreducible error and doesn't matter how well the fit is.

Provide a sketch of typical (squared) bias, variance, training er- ror, test error, and Bayes (or irreducible) error curves, on a sin- gle plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.  
2. Explain why each of the five curves has the shape displayed in part (a).

**Solution.**  

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
import logging 

def squared_bias(x):
    return .002*(-x+10)**3

def variance(x):
    return .002*x**3 

def training_error(x):
    return 2.38936 - 0.825077*x + 0.176655*x**2 - 0.0182319*x**3 + 0.00067091*x**4

def test_error(x):
    return 3 - 0.6*x + .06*x**2

def bayes_error(x):
    return x + 1 - x

x = np.arange(0.0, 10.0, 0.02)

# suppress missing font warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    logging.getLogger('matplotlib.font_manager').setLevel(logging.ERROR)

    plt.xkcd()
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))

    ax.plot(x, squared_bias(x), label='squared bias')
    ax.plot(x, variance(x), label='variance')
    ax.plot(x, training_error(x), label='training error')
    ax.plot(x, test_error(x), label='test error')
    ax.plot(x, bayes_error(x), label='Bayes error')
    
    ax.legend(loc='upper center')
    ax.set_xlabel('model flexibility')

    ax.set_xticklabels([])
    ax.set_yticklabels([])

    plt.show()
```

2. Squared bias typically falls as flexibility increases due to the model being better fit to approximate the true functional form of our data. Variance will rise as flexibility increases as a flexible model changes values more drastically responding to fluctuations and noise in the observed data, whereas an inflexible model does not. Training error falls as model flexibility increases as a flexible model can overlearn on the training data to approximate it perfectly. Test error increases for the same reason. The irreducible error is constant irrespective of model flexibility as it will always exist independently of our model.

**Question 5.**  
What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?  

**Solution.**  
The advantages of a flexible approach for regression or classification is that you are less prone to overlearning and can make accurate extrapolations even when large quantities of data are not available. Furthermore, less flexible models are more straightforward in their interpretability and are highly useful in inference. Flexible approaches are much more powerful when performing predictions with large quantities of data, and model interpretability is of secondary importance.  

**Question 6.**  

Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a para- metric approach to regression or classification (as opposed to a non- parametric approach)? What are its disadvantages?

**Solution.**  
The advantages of a parametric approach are that in cases where there is good reason to assume an underlying parametric form of the data, parametric analyses require less data and are less influenced by noise in the data. The advantages of a nonparametric approach is that previous assumptions on the true parametric forms of the data need not be made, and the disadvantages are that they are more error prone due to their easily being swayed by noise. 

**Question 7.**  

The table below provides a training data set containing six observa- tions, three predictors, and one qualitative response variable.

| Obs. | $X_1$ | $X_2$ | $X_3$ | $Y$ |  
|-|-|-|-|-|  
| 1 | 0 | 3 | 0 | Red|  
|2 | 2 | 0 | 0 | Red |  
| 3 | 0 | 1 | 3 | Red |  
| 4 | 0 | 1 | 2 | Green |  
| 5 | -1 | 0 | 1 | Green |  
| 6 | 1 | 1 | 1 | Red |   

Suppose we wish to use this dataset to make a prediction for $Y$ when $X_1 = X_2 = X_3 = 0$ using $K$-nearest neighbors.  

1. Compute the Euclidean distance between each observation and thetestpoint,X1 =X2 =X3 =0.  
2. What is our prediction with K = 1? Why?  
3. What is our prediction with K = 3? Why?  
4. If the Bayes decision boundary in this problem is highly non- linear, then would we expect the best value for K to be large or small? Why?  

```python
import pandas as pd 

d = {'X1': pd.Series([0,2,0,0,-1,1]),
     'X2': pd.Series([3,0,1,1,0,1]),
     'X3': pd.Series([0,0,3,2,1,1]),
     'Y': pd.Series(['Red','Red','Red','Green','Green','Red'])}
df = pd.DataFrame(d)
df.index = np.arange(1, len(df) + 1)
df
```

```python
# 1. euclidean distance from test point X_1 = X_2 = X_3 = 0

squared = df[["X1", "X2", "X3"]] ** 2
distance = np.sqrt(np.sum(squared, axis = 1))
df['distance'] = distance
df_ordered = df.sort_values(by="distance")
df_ordered
```

```python
# 2. prediction with K = 1 

from collections import Counter

counter = Counter(df_ordered.iloc[:1]['Y'])
mode_value, mode_count = counter.most_common(1)[0]

print(mode_value)
```

```python
# 3. prediction with K = 3

counter = Counter(df_ordered.iloc[:3]['Y'])
mode_value, mode_count = counter.most_common(1)[0]

print(mode_value)
```

4. Larger values of K tend to smooth out the decision boundary, as they are less influenced by one close point and take the value of the majority of a larger neighborhood. Thus, in the case where we want to classify a point that lies just within group A of a portion of a non linear decision boundary that peeks into group B, a large value of K is more likely to count the neighbors in group B and classify the point as belonging to group B.
